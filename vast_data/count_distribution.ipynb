```python
#few_train zero_train zero_dev zero_test
def count(filename,step):
    pro_tag='1'
    con_tag='0'
    neu_tag='2'
    #filename='few_vast_train.raw'
    targets=set()
    f=open(filename,'r',encoding='utf-8-sig')
    lines=f.readlines()
    label_dict={}
    for i in range(0,len(lines),step):
        
        target=lines[i+1].strip()
        targets.add(target)
        label=lines[i+2].strip()
        #print(type(label))
        if label not in label_dict:
            label_dict[label]=1
        else:
            label_dict[label]+=1
    print('pro_tag:',label_dict['1'])
    print('con_tag:',label_dict['0'])
    print('neu_tag:',label_dict['2'])
    return targets
```
